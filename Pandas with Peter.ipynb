```python
election_data = pd.read_csv('trump_ratings_first400 (1).csv')
election_data.head()
```

```python
#params = np.polyfit(x,y,1)
#function = np.poly1d(params)
#expected_price = function(x)
x = election_data['datenumber']
y_1 = election_data['adjusted_approve']
y_2 = election_data['adjusted_disapprove']
parameters = np.polyfit(x,y_1,2)
function_1 = np.poly1d(parameters)
expected_approval = function_1(x)
plt.plot_date(x,y_1)1
plt.plot_date(x,expected_approval)
```

```python
parameters = np.polyfit(x,y_2,2)
function_2 = np.poly1d(parameters)
expected_disapproval = function_2(x)
plt.plot_date(x,y_2)
plt.plot_date(x,expected_disapproval)
plt.plot_date(x,y_1)
plt.plot_date(x,expected_approval)
```

```python
last = x.tail(1)
new = np.arange(last, last + 500,1)
x = x.append(pd.DataFrame(new))
```

```python
predicted_approval = function_1(x)
predicted_disapproval = function_2(x)
plt.plot_date(x, predicted_approval)
plt.plot_date(x, predicted_disapproval)
plt.plot_date(election_data['datenumber'],y_1)
plt.plot_date(election_data['datenumber'],y_2)
```
